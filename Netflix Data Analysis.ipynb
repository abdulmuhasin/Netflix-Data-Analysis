From column Date Added all date values are not in the same format

```python
# import the parser library from dateutil to set the datatime in same format
from dateutil import parser
```

```python
def same_parse_date(date):
    if pd.isnull(date):
        return pd.NaT
    try:
        return parser.parse(date, dayfirst=False)
    except (ValueError, TypeError):
        return pd.NaT  # Return NaT if parsing fails
```

```python
# Apply the function to date columns
df['Date Added'] = df['Date Added'].apply(same_parse_date)
```

# Top 10 Genres

```python
# Split the 'Listed In' column and explode into individual rows
df['Genres'] = df['Listed In'].str.split(', ')
genre_explode = df.explode('Genres')
```

```python
# Count the top 10 genres
top_genres = genre_explode['Genres'].value_counts().head(10)
top_genres
```

```python
# Plotting
plt.figure(figsize=(10, 6))
top_genres.plot(kind='bar', color='#3498DB')
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Genre')
plt.ylabel('Count of Titles')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by 'Type' and 'Genres'
type_genre_counts = genre_explode.groupby(['Type', 'Genres']).size().unstack(fill_value=0)
```

```python
# Plotting the top 5 genres for each type
type_genre_counts.T.head(5).plot(kind='bar', figsize=(12, 6), colormap='Set3')
plt.title('Top Genres by Content Type (Movies vs TV Shows)')
plt.xlabel('Genre')
plt.ylabel('Count of Titles')
plt.xticks(rotation=45)
plt.legend(title='Content Type')
plt.show()
```

```python
# Group by Year Added and Genres
genre_year = genre_explode.groupby([df['Year Added'], 'Genres']).size().unstack(fill_value=0)
```

```python
# Plotting the trend for top 3 genres
genre_year[top_genres.index[:3]].plot(kind='line', marker='o', figsize=(12, 6))
plt.title('Trend of Top 3 Genres Over the Years')
plt.xlabel('Year')
plt.ylabel('Count of Titles')
plt.grid(True)
plt.legend(title='Genres')
plt.show()
```

# Genre By Rating

```python
# Group by Rating and Genres
rating_genre_counts = genre_explode.groupby(['Rating', 'Genres']).size().unstack(fill_value=0)
```

```python
# Plot the top 5 genres for each rating
rating_genre_counts.T.head(5).plot(kind='bar', stacked=True, figsize=(12, 6), colormap='viridis')
plt.title('Genre Distribution by Rating')
plt.xlabel('Genre')
plt.ylabel('Count of Titles')
plt.legend(title='Rating', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

# Top 10 Directors with Rating

```python
# Group by Director and Rating, and count the number of titles
director_rating_counts = df.groupby(['Director', 'Rating']).size().reset_index(name='Count')
```

```python
# Get the top 10 directors by the total number of titles
top_directors = df['Director'].value_counts().head(10).index
```

```python
# Filter the grouped data to keep only the top 10 directors
top_director_ratings = director_rating_counts[director_rating_counts['Director'].isin(top_directors)]
```

```python
# Plotting
plt.figure(figsize=(12, 6))
for director in top_directors:
    data = top_director_ratings[top_director_ratings['Director'] == director]
    plt.bar(data['Rating'], data['Count'], label=director)

plt.title('Top 10 Directors with their Ratings')
plt.xlabel('Rating')
plt.ylabel('Count of Titles')
plt.xticks(rotation=45)
plt.legend(title='Director', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
